## Command Line Solutions

### 1. Download the `students.csv` file

To download the `students.csv` file from the provided GitHub URL, use the `wget` command:


```{bash}
which wget
rm -f students.csv
/opt/homebrew/bin/wget https://raw.githubusercontent.com/stephaniehicks/jhustatprogramming2024/main/projects/01-project/students.csv
```


### 2. Display the contents of the `students.csv` file using the `cat` command


```{bash}
cat students.csv
```


### 3. Display only the first 5 lines of the file using `head` to show the first few lines


```{bash}
head -n 5 students.csv
```


### 4. Display only the last 3 lines of the file using `tail` to show the last few lines


```{bash}
tail -n 3 students.csv
```


### 5. Count the number of lines in the file using the `wc` command


```{bash}
wc -l students.csv
```


### 6. Find all students who are taking “Math” as a subject using `grep`


```{bash}
grep "Math" students.csv
```


### 7. Find all female students using `grep` or `awk` to filter based on gender

Using `grep`:


```{bash}
grep ",F," students.csv
```


### 8. Sort the file by the students’ ages in ascending order using `sort`


```{bash}
sort -t "," -k3,3n students.csv
```


### 9. Find the unique subjects listed in the file using `cut`, `sort`, and `uniq`


```{bash}
cut -d "," -f6 students.csv | sort | uniq
```


### 10. Calculate the average grade of the students using `awk`


```{bash}
awk -F "," '{sum+=$5} END {print "Average Grade:", sum/(NR-1)}' students.csv
```


### 11. Replace all occurrences of “Math” with “Mathematics” in the file using `sed`

To make an in-place replacement:


```{bash}
sed -i '' 's/Math/Mathematics/g' students.csv
```


### 12. Display the final `students.csv` file

```{bash}
cat students.csv
```
